```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('output.csv')
# histogram of the num2crit
plot_blocks = df.num2crit.plot(kind='hist', color='purple', edgecolor='black', figsize=(10,7))
plt.title('Distribution of # of blocks needed to meet criterion', size=18)
plt.xlabel('Number of training blocks', size=18)
plt.ylabel('Frequency', size=18)
plt.show(plot_blocks)

# histogram of the novel cat correct in p2
plot_noevelCat = df.total_novel_cat.plot(kind='hist', color='purple', edgecolor='black', figsize=(10,7))
plt.title('Distribution of # of novel categories answered correctly in P2', size=18)
plt.xlabel('Number of novel categories', size=18)
plt.ylabel('Frequency', size=18)
plt.show(plot_noevelCat)

```

```python
import seaborn as sns

# regression plot using seaborn
df_Discriminative = df[df['condition'] == 1]
df_Generative = df[df['condition'] == 2]

fig = plt.figure(figsize=(10, 7))
sns.regplot(x=df_Discriminative.delta_p2b2, y=df_Discriminative.delta_p2b1, color='blue', marker='+', label= 'Discriminative')
sns.regplot(x=df_Generative.delta_p2b2, y=df_Generative.delta_p2b1, color='green', marker='+', label= 'Generative')

# Legend, title and labels.
plt.legend(loc="upper left")
plt.title('Relation amongst # of correct delta categories in block 1, # of correct delta categories in block 2, and learning mode', size=14)
plt.xlabel('Number of correct delta categories in block 1', size=12)
plt.ylabel('Number of correct delta categories in block 2', size=12)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('metCritpnums.csv')


#get condition 1 pnums
cond1_pnums = []
cond1_filter = np.where(df['condition']== 1, df['pnum'], 0)
cond1_filter_collected = np.unique(cond1_filter)

#print(cond1_filter_collected)

df_cond1_filter_collected = pd.DataFrame(cond1_filter_collected, columns = ['pnums'])
#print(df_cond1_filter_collected)
df_cond1_filter_collected = df_cond1_filter_collected.drop(index= 0)
#print(df_cond1_filter_collected)

for subjects in df_cond1_filter_collected['pnums'].unique():
    cond1_pnums.append(subjects)

print('cond1 pnums:')
print(cond1_pnums)




#get condition 2 pnums
cond2_pnums = []
cond2_filter = np.where(df['condition']== 2, df['pnum'], 0)
cond2_filter_collected = np.unique(cond2_filter)

#print(cond2_filter_collected)

df_cond2_filter_collected = pd.DataFrame(cond2_filter_collected, columns = ['pnums'])
#print(df_cond2_filter_collected)
df_cond2_filter_collected = df_cond2_filter_collected.drop(index= 0)
#print(df_cond2_filter_collected)

for subjects in df_cond2_filter_collected['pnums'].unique():
    cond2_pnums.append(subjects)

print('cond2 pnums')
print(cond2_pnums)



#make a cond1_df by dropping cond2 and cond3 pnums
#this loop drops cond2_pnums
cond1_df = df
for item in cond2_pnums:
    cond1_df = cond1_df.drop(index = cond1_df[cond1_df['pnum'] == item].index)
    
#save cond1_df as a csv    
cond1_df.to_csv( "C:/Users/apers/Downloads/GloblocData/discriminative_data.csv", index=False, encoding='utf-8-sig')


cond2_df = df
for item in cond1_pnums:
    cond2_df = cond2_df.drop(index = cond2_df[cond2_df['pnum'] == item].index)


cond2_df.to_csv( "C:/Users/apers/Downloads/GloblocData/generative_data.csv", index=False, encoding='utf-8-sig')


        

```
